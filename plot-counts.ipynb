```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dir = "data"
save = True
```

# Get Data

```python
df = pd.read_csv(f'{data_dir}/station-counts.csv', index_col='datetime')
```

# Plot Histogram

```python
# get data
bar_df = pd.DataFrame(df.sum().sort_values(ascending=False)[9:24], columns=['num_people'])

x = bar_df['num_people'][::-1]/52
y = bar_df.index[::-1]

# plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.barh(y, x, align='center', capsize=3)
ax.xaxis.grid(True)
ax.set_yticks(y)
ax.set_xlabel('Number of People Per Week', fontsize=12)
ax.set_title('Top 15 Stations by Number of People per Week', fontsize=16)

# save
if save:
    plt.savefig("figures/station-count-week.png",
                bbox_inches="tight",
                transparent=False,)
```

# Plot Line Chart for One Week

```python
df[(df.index >= '2018-03-24 00:00:00')
   & (df.index <= '2018-03-31 00:00:00')][['34 ST-HERALD SQ',
                                           'TIMES SQ-42 ST',
                                           '1 AV',
                                           'LEXINGTON AV/63',
                                           '42 ST-PORT AUTH']].plot(ylim = (0, 3500), figsize=(12, 8))

plt.xticks([0, 6, 12, 18, 24, 30, 36], ['Sat', 'Sun', 'Mon', 'Tues', 'Wed', 'Thurs', 'Fri'])
plt.xlabel('Day of the Week', fontsize=16)
plt.ylabel('Number of People', fontsize=16)
plt.title('One Week of Station Traffic for the Top 5 Stations', fontsize=24)

if save:
    plt.savefig("figures/day-count-20200324_20200331.png",
                bbox_inches="tight",)
```
